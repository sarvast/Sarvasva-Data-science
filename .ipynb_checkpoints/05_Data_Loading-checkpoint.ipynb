### Step 1: Import Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Ready to load data! ðŸšš")
```

### Step 2: Load CSV File
`pd.read_csv()` function use hota hai.

```python
# Path sahi dena zaroori hai
df = pd.read_csv('Datasets/house_prices.csv')

# 'df' ka matlab DataFrame (Table)
print("Data Load Ho Gaya!")
```

### Step 3: Inspect Data (Data ko taado)
Pura data print mat karo, bas upar ki 5 lines dekho.

```python
df.head() # Top 5 rows
```

### Step 4: Statistics (Data ki kundli)
`describe()` function sab kuch bata deta hai (Mean, Max, Min).

```python
df.describe()
```

### Step 5: Visualize Relationships
Kya Area badhne se Price badh raha hai?

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Area_SqFt', y='Price_Lakhs', data=df, color='green')
plt.title("Area vs Price")
plt.show()
```
