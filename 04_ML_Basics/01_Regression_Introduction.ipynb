# ðŸ¤– Machine Learning 101: Linear Regression

**Goal**: Aaj hum Computer ko sikhayenge ki **Data se Prediction** kaise karte hain.

**Problem**: Humare paas `Experience` (Years) hai, humein `Salary` predict karni hai.
Jitna zyada experience, utni zyada salary. Computer ko ye pattern khud dhoondna hai.

### Step 1: Libraries Import

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

print("Setup Complete!")
```

### Step 2: Create Dummy Data
Asli data nahi hai, toh khud bana lete hain.
- **X**: Experience (Years)
- **y**: Salary (Thousand Rupees)

```python
# Input (Experience) - Isse 2D array hona zaroori hai isliye reshape(-1, 1) kiya
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)

# Output (Salary)
y = np.array([15, 25, 35, 45, 55]) # 10k ka increment har saal

# Data ko dekhte hain
plt.scatter(X, y, color='blue')
plt.xlabel("Experience (Years)")
plt.ylabel("Salary (k)")
plt.title("Raw Data")
plt.show()
```

### Step 3: Train the Model
Ab hum `LinearRegression` model banayenge aur usse bolenge: "Beta, X aur y ko study karo aur pattern seekho" (`fit` method).

```python
model = LinearRegression()

# Training (Learning)
model.fit(X, y)

print("Model Train ho gaya! ðŸŽ“")
```

### Step 4: Testing (Prediction)
Ab dekhte hain model kitna hoshiyar hai.
Hum puchenge: "Agar **7 saal** ka experience hai, toh salary kya honi chahiye?"

```python
new_exp = np.array([[7]]) # 7 years

prediction = model.predict(new_exp)

print(f"Prediction for 7 years: {prediction[0]} k")
# Answer 75 aana chahiye agar pattern sahi pakda hai toh
```

### Step 5: Visualization Line
Model ne asli mein kya seekha? Wo ek seedhi line (Linear) kheechta hai jo sabse best fit ho.

```python
plt.scatter(X, y, color='blue') # Original dots
plt.plot(X, model.predict(X), color='red') # Machine ki line
plt.title("Machine Learning Fit")
plt.show()
```

## ðŸ§  Kya hua andar?
Computer ne Formula dhoond liya: `Salary = 10 * Experience + 5`
- `Coef (m)`: 10
- `Intercept (c)`: 5

```python
print("Coefficient (Slope):", model.coef_)
print("Intercept:", model.intercept_)
```
